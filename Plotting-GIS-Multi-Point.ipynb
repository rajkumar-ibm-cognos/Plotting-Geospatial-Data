```python
#Plotting of Geometrical Shape known as "Multi-Point"
import matplotlib.pyplot as plt

# Define your data points
x_values = [1, 2, 3, 4, 5]
y_values = [2, 3, 5, 7, 6]

# Plot the points
plt.scatter(x_values, y_values, color='blue', marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multi-Point Graph')

# Show the plot
plt.grid(True)
plt.show()
```
