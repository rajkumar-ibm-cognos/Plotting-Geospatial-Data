```python
#Plotting of Geometrical Shape known as "Polygon"
import matplotlib.pyplot as plt

# Define the coordinates of the vertices of the polygon
x = [1, 2, 3, 2, 1]
y = [1, 2, 1, 0, 1]

# Plot the polygon
plt.figure()
plt.plot(x, y, marker='o')  # 'o' for markers at vertices
plt.fill(x, y, alpha=0.3)    # Fill the polygon with color and transparency
plt.title('Polygon Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axis('equal')  # Set equal scaling for both axes
plt.show()
```
