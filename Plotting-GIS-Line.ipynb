```python
#Plotting of Geometrical Shape known as "Line"
import matplotlib.pyplot as plt

# Define the coordinates of the points
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 6]

# Plot the line
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot of a Line')

# Show the plot
plt.show()
```
