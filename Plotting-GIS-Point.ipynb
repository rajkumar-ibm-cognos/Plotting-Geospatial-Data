```python
#Plotting of Geometrical Shape known as "Point"
import matplotlib.pyplot as plt
# Coordinates of the point
x = 2
y = 3
# Plotting the point
plt.scatter(x, y, color='brown', marker='o')  # 'o' is the marker style for a circle
# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting a Point')
# Displaying the plot
plt.grid(True)  # Adding grid
plt.show()
```
